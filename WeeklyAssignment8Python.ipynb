```python
import csv
from collections import defaultdict
from collections import Counter

def make_data_point(line, header):
    output = {}
    for index, header_item in enumerate(header):
        output[header_item] = line[index]
    return output

with open("denver-1.nov.csv") as inf:
    reader = csv.reader(inf)
    header = next(reader)
    police_data = []
    for line in reader:
        police_data.append(make_data_point(line, header))
```

In this data I want to look at the different outcomes of a traffic stop

```python
police_data
```

I decided to aggregate the data by outcome type. 

```python
outcomeCounter = defaultdict(list)
for i in police_data:
    outcomeCounter[i['outcome']].append(i)
```

```python
outcomeCounter
```

```python
for i in outcomeCounter:
    print(i)
```

I found that these are the four types of outcomes at a traffic stop. Now I want to find out how many of each action is represented in this data.

```python
outcomeList = []
for i in outcomeCounter:
    outcomeCounter2 = outcomeCounter[i]
    for x in outcomeCounter2:
        outcomeList.append(x['outcome'])
#These are the four actions that can happen as an outcome of the traffic stop. 
```

```python
Counter(outcomeList)
```

Above is the data I will use to put in my chart.js chart. This is the statistic on whether a stop will end up as a 'NA', 'warning', 'citation', or 'arrest'. 

We have gotten the 4 outcomes and the total traffic stops that resulted in each. From here I want to look at the precincts that have the most 'NA's. I want to see which precincts are having the most stops, but are not citing, warning, or arrestting the person. This is not included in my chart.js.

```python
precinctNaList = []
for i in outcomeCounter['NA']:
    precinctNaList.append(i['precinct'])
```
